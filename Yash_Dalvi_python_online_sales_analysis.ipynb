```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import io
```

```python
df = pd.read_csv(io.BytesIO(upload["python graphs.csv"]))
```

```python
df.head()
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('â‚¹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].str.replace('â‚¹', '').str.replace(',', '').astype(float)
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(
    data=df,
    x='rating',
    hue='rating',
    palette='pastel',
    order=df['rating'].value_counts().index,
    legend=False
)
plt.title('Rating Distribution')
plt.ylabel('')
plt.show()
```

```python
df['main_category'] = df['category'].str.split('|').str[0]
avg_price = df.groupby('main_category')['discounted_price'].mean().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    x=avg_price.values,
    y=avg_price.index,
    hue=avg_price.index,
    palette='viridis',
    legend=False
)
plt.title('Top 10 Categories by Avg Discounted Price')
plt.xlabel('Average Discounted Price')
plt.ylabel('Main Category')
plt.show()
```

```python
avg_price = df.groupby('main_category')['discounted_price'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(
    x=avg_price.values,
    y=avg_price.index,
    hue=avg_price.index,
    palette='coolwarm',
    legend=False
)
plt.title('Average Discounted Price by Category')
plt.xlabel('Average Price (₹)')
plt.ylabel('Category')
plt.grid(True, axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
top_rated = df[['product_name', 'rating_count']].dropna().sort_values(by='rating_count', ascending=False).head(15)
fig = px.treemap(top_rated, path=['product_name'], values='rating_count',
                 title='Top Products by Rating Count')
fig.show()
```

```python
plt.figure(figsize=(12, 6))

top_cats = df['main_category'].value_counts().nlargest(6).index

sns.boxplot(
    data=df[df['main_category'].isin(top_cats)],
    x='main_category',
    y='discounted_price',
    hue='main_category',
    palette='Set2',
    dodge=False,
    legend=False
)

plt.title('Price Variability in Top 6 Categories')
plt.xlabel('Main Category')
plt.ylabel('Discounted Price')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
cat_counts = df['main_category'].value_counts().nlargest(10)
colors = plt.cm.Accent.colors[:10]  # Use Accent colormap
plt.bar(cat_counts.index, cat_counts.values, color=colors)
plt.title('Top 10 Product Categories by Count')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
top_cats = df['main_category'].value_counts().nlargest(5).index
plt.figure(figsize=(12, 6))
sns.violinplot(data=df[df['main_category'].isin(top_cats)],
               x='main_category', y='rating', palette='Set2')
plt.title('Rating Distribution in Top 5 Categories')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df[['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].corr(),
            annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Between Numeric Features")
plt.tight_layout()
plt.show()
```

```python
bins = [0, 10, 100, 500, 1000, 5000, 10000]
df['rating_count_bin'] = pd.cut(df['rating_count'], bins)
plt.figure(figsize=(10, 5))
sns.countplot(
    x='rating_count_bin',
    data=df,
    hue='rating_count_bin',
    palette='cool',
    dodge=False,
    legend=False
)

plt.title("Distribution of Products by Rating Count Groups")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top10 = df[['product_name', 'rating_count']].dropna().sort_values(by='rating_count', ascending=False).head(10)
top10['short_name'] = top10['product_name'].apply(lambda x: ' '.join(str(x).split()[:3]))
plt.figure(figsize=(10, 6))
plt.hlines(y=top10['short_name'], xmin=0, xmax=top10['rating_count'], color='green')
plt.plot(top10['rating_count'], top10['short_name'], "o")
plt.title("Top 10 Products by Rating Count (Lollipop Style)")
plt.xlabel("Rating Count")
plt.tight_layout()
plt.show()
```

```python
df['discount_level'] = ['High' if x >= 50 else 'Low' for x in df['discount_percentage'].fillna(0)]
plt.figure(figsize=(10, 5))
sns.kdeplot(data=df[df['discount_level'] == 'High']['rating'].dropna(), label='High Discount', fill=True)
sns.kdeplot(data=df[df['discount_level'] == 'Low']['rating'].dropna(), label='Low Discount', fill=True)
plt.title("Rating Distribution for High vs Low Discounts")
plt.xlabel("Rating")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.stripplot(data=df, x='rating', y='discounted_price', hue='main_category', dodge=True, jitter=True, alpha=0.6)
plt.title("Discounted Price vs Rating by Category")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
